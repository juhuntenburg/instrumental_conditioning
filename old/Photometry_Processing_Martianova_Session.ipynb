```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.ndimage import gaussian_filter1d
from sklearn.linear_model import LinearRegression
from scipy.special import legendre
from photometry_functions import airPLS, get_zdFF
```

```python
sns.set_style('ticks')
sns.set_context('talk')
import warnings; warnings.simplefilter('ignore')
```

```python
data_dir = '/home/julia/data/ict/'
```

### Load one session and concatenate all trials

```python
mouse = 'UUU'
df = pd.read_pickle(data_dir+'mice/{}.pkl'.format(mouse))
```

```python
day=0
signal_series = df[(df['day']==day)]['gpmt']
reference_series = df[(df['day']==day)]['rpmt']
```

```python
raw_signal = np.empty(0)
for trial in signal_series:
    raw_signal = np.append(raw_signal, trial)
    
raw_reference = np.empty(0)
for trial in reference_series:
    raw_reference = np.append(raw_reference, trial)
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax1.plot(raw_signal,'blue',linewidth=0.5)
ax2 = fig.add_subplot(212)
ax2.plot(raw_reference,'purple',linewidth=0.5)
plt.tight_layout()
sns.despine()
```

### Smooth

```python
smooth_signal = gaussian_filter1d(raw_signal, sigma=100, output=np.float64)
smooth_reference = gaussian_filter1d(raw_reference, sigma=100, output=np.float64)
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax1.plot(smooth_signal,'blue',linewidth=1.5)
ax2 = fig.add_subplot(212)
ax2.plot(smooth_reference,'purple',linewidth=1.5)
plt.tight_layout()
sns.despine()
```

### Linear regression to fit baseline

```python
predict_signal = airPLS(smooth_signal, lambda_=5e8, porder=1, itermax=50)
predict_reference = airPLS(smooth_reference, lambda_=5e8, porder=1, itermax=50)
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax1.plot(smooth_signal,'blue',linewidth=1.5)
ax1.plot(predict_signal,'black',linewidth=1.5)
ax2 = fig.add_subplot(212)
ax2.plot(smooth_reference,'purple',linewidth=1.5)
ax2.plot(predict_reference,'black',linewidth=1.5)
plt.tight_layout()
sns.despine()
```

### Remove baseline

```python
signal = smooth_signal - predict_signal 
reference = smooth_reference - predict_reference
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(211)
ax1.plot(signal,'blue',linewidth=1.5)
ax2 = fig.add_subplot(212)
ax2.plot(reference,'purple',linewidth=1.5)
plt.tight_layout()
sns.despine()
```

### Standardize signal

```python
z_reference = (reference - np.median(reference)) / np.std(reference)
z_signal = (signal - np.median(signal)) / np.std(signal)
```

```python
fig = plt.figure(figsize=(8, 4))
ax1 = fig.add_subplot(121)
ax1.plot(z_signal,'blue',linewidth=1.5)
ax2 = fig.add_subplot(122)
ax2.plot(z_reference,'purple',linewidth=1.5)
plt.tight_layout()
sns.despine()
```
